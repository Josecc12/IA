```python
from matplotlib import pyplot as plt 
from skimage import color
import cv2
import common
import pylab
```

```python
archivo = 'auto.jpg'
```

```python
auto_image = cv2.imread(archivo)
plt.imshow(cv2.cvtColor(auto_image,cv2.COLOR_BGR2RGB))
```

```python
auto_gris = cv2.cvtColor(auto_image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(auto_gris,cmap='gray')
```

```python
sobelXa = cv2.Sobel(auto_gris,cv2.CV_64F,1,0,ksize=9)
plt.imshow(sobelXa,cmap='gray')
```

```python
sobelYa = cv2.Sobel(auto_gris,cv2.CV_64F,0,1,ksize=9)
plt.imshow(sobelYa,cmap='gray')
```

```python
sobelXY = cv2.Sobel(auto_gris,cv2.CV_64F,1,1,ksize=9)
plt.imshow(sobelXY,cmap='gray')
```
