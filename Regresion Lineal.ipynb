###  1.Linear Regression

```python
import numpy as np
import seaborn as sns
import scipy as stats
import statsmodels.api as sm
```

```python
auto = sns.load_dataset('mpg')
```

```python
auto
```

### 1.1 Regression Model
y=mx+b

```python
x= auto['weight']
x = sm.add_constant(x)
y = auto ['mpg'] 
```

```python
result = sm.OLS(y,x).fit()
```

```python
result.summary()
```

```python
result.params
```

```python
result.rsquared
```

## 1.2 Grafic Linear Regression

```python
sns.set_theme(color_codes=True)
```

```python

ax1=sns.regplot(data=auto,x='weight', y='mpg')
ax1.set_xlabel('Peso')
ax1.set_ylabel('MPG')
ax1.set_title('Relacion Peso-MPG')
ax1.plot()
```

```python
ax2= sns.jointplot(x='weight' , y='mpg' , data=auto , kind='reg' , truncate=False , color='r', height=7)
```
